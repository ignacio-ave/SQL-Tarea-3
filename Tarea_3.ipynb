```
CREATE TABLE Clientes
(
  Nombre VARCHAR(50) NOT NULL,
  Persona_de_contacto VARCHAR(50) NOT NULL,
  Correo_de_contacto VARCHAR(50) NOT NULL,
  Telefono INT NOT NULL,
  Domicilio VARCHAR(50) NOT NULL,
  Giro VARCHAR(50) NOT NULL,
  Rut_empresa INT NOT NULL,
  PRIMARY KEY (Rut_empresa)
);

CREATE TABLE Ingeniero
(
  Rut_ingeniero INT NOT NULL,
  Nombre VARCHAR(50) NOT NULL,
  Domicilio VARCHAR(50) NOT NULL,
  Telefono INT NOT NULL,
  Sueldo FLOAT NOT NULL,
  Banco VARCHAR(50) NOT NULL,
  Numero_de_cuenta INT NOT NULL,
  PRIMARY KEY (Rut_ingeniero)
);

CREATE TABLE Proyecto
(
  Codigo INT NOT NULL,
  Descripcion VARCHAR(50) NOT NULL,
  Valor_Proyecto FLOAT NOT NULL,
  Duracion INT NOT NULL,
  Inicio DATE NOT NULL,
  Fin DATE NOT NULL,
  Rut_empresa INT NOT NULL,
  PRIMARY KEY (Codigo),
  FOREIGN KEY (Rut_empresa) REFERENCES Clientes(Rut_empresa)
);

CREATE TABLE Planilla
(
  Horas_Semanales INT NOT NULL,
  Codigo INT NOT NULL,
  Rut_ingeniero INT NOT NULL,
  FOREIGN KEY (Codigo) REFERENCES Proyecto(Codigo),
  FOREIGN KEY (Rut_ingeniero) REFERENCES Ingeniero(Rut_ingeniero)
);


```

````python
import pandas as pd
import sqlite3
from faker import Faker
import random
from datetime import datetime, timedelta

# Crear una instancia de Faker
fake = Faker()

# Conectar a la base de datos
conn = sqlite3.connect('Proyecto.db')
cursor = conn.cursor()

# Función para generar datos para la tabla Clientes
def generar_datos_clientes(n):
    clientes = []
    for _ in range(n):
        nombre = fake.company()
        persona_contacto = fake.name()
        correo_contacto = fake.email()
        telefono = fake.random_number(digits=8, fix_len=True)
        domicilio = fake.address().replace('\n', ', ')
        giro = fake.bs()
        rut_empresa = fake.random_number(digits=9, fix_len=True)
        clientes.append((nombre, persona_contacto, correo_contacto, telefono, domicilio, giro, rut_empresa))
    return clientes

# Función para generar datos para la tabla Ingeniero
def generar_datos_ingenieros(n):
    ingenieros = []
    for _ in range(n):
        rut_ingeniero = fake.random_number(digits=9, fix_len=True)
        nombre = fake.name()
        domicilio = fake.address().replace('\n', ', ')
        telefono = fake.random_number(digits=8, fix_len=True)
        sueldo = round(random.uniform(1000, 10000), 2)
        banco = fake.company()
        numero_cuenta = fake.random_number(digits=10, fix_len=True)
        ingenieros.append((rut_ingeniero, nombre, domicilio, telefono, sueldo, banco, numero_cuenta))
    return ingenieros

# Función para generar datos para la tabla Proyecto
def generar_datos_proyectos(n, rut_empresas):
    proyectos = []
    for _ in range(n):
        codigo = fake.random_number(digits=5, fix_len=True)
        descripcion = fake.catch_phrase()
        valor_proyecto = round(random.uniform(50000, 500000), 2)
        duracion = random.randint(1, 24)
        inicio = fake.date_between(start_date='-2y', end_date='today')
        fin = inicio + timedelta(days=duracion*30)
        rut_empresa = random.choice(rut_empresas)
        proyectos.append((codigo, descripcion, valor_proyecto, duracion, inicio, fin, rut_empresa))
    return proyectos

# Función para generar datos para la tabla Planilla
def generar_datos_planilla(n, codigos_proyectos, rut_ingenieros):
    planillas = []
    for _ in range(n):
        horas_semanales = random.randint(20, 40)
        codigo = random.choice(codigos_proyectos)
        rut_ingeniero = random.choice(rut_ingenieros)
        planillas.append((horas_semanales, codigo, rut_ingeniero))
    return planillas

# Generar datos
clientes = generar_datos_clientes(50)
ingenieros = generar_datos_ingenieros(50)
rut_empresas = [cliente[6] for cliente in clientes]
proyectos = generar_datos_proyectos(50, rut_empresas)
codigos_proyectos = [proyecto[0] for proyecto in proyectos]
rut_ingenieros = [ingeniero[0] for ingeniero in ingenieros]
planillas = generar_datos_planilla(100, codigos_proyectos, rut_ingenieros)

# Insertar datos en la base de datos
cursor.executemany('INSERT INTO Clientes VALUES (?, ?, ?, ?, ?, ?, ?)', clientes)
cursor.executemany('INSERT INTO Ingeniero VALUES (?, ?, ?, ?, ?, ?, ?)', ingenieros)
cursor.executemany('INSERT INTO Proyecto VALUES (?, ?, ?, ?, ?, ?, ?)', proyectos)
cursor.executemany('INSERT INTO Planilla VALUES (?, ?, ?)', planillas)

conn.commit()

````

## **Consulta (b)**

Debido a que la mayoria de atributos son categoricos el describe puede que no nos de una parte de la informacion como la distribucion estandar pero podemos ingeniarnosla con la entropia del atributo



Índices y Entropía

    Alta Entropía:
        Descripción: Los valores del atributo están distribuidos de manera más uniforme y hay una mayor variedad de categorías.
        Ventajas: Los índices en atributos de alta entropía tienden a ser más selectivos, lo que puede mejorar el rendimiento de las consultas que buscan valores específicos. Esto se debe a que cada valor tiene una menor probabilidad de repetirse.
        Desventajas: Si el número de valores únicos es extremadamente alto, el tamaño del índice también puede ser grande, lo que puede aumentar el costo de mantenimiento del índice.

    Baja Entropía:
        Descripción: Unas pocas categorías dominan la distribución del atributo.
        Ventajas: Los índices en atributos de baja entropía pueden ser útiles para consultas que buscan las categorías más frecuentes. Estos índices suelen ser más pequeños y menos costosos de mantener.
        Desventajas: Los índices en atributos de baja entropía pueden ser menos selectivos, lo que puede reducir el rendimiento de las consultas que buscan valores específicos, ya que muchas filas pueden tener el mismo valor del atributo.
